```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime
from scipy.interpolate import interp1d

from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
import matplotlib.dates as mdates
```

```python
# add horizontal lines
fineTune = False

x1 = pd.date_range(start = datetime(2020, 4, 7, 6,   0), 
                   end   = datetime(2020, 4, 7, 19, 30), 
                   freq  = '45min')

x2 = pd.date_range(start = datetime(2020, 4, 20, 6,   0), 
                   end   = datetime(2020, 4, 20, 19, 30), 
                   freq  = '45min')

# April 7th data
y1_95 = np.array([0, 0, 20,    np.nan,  np.nan,     655,  np.nan,   805,  902,  np.nan,  1003, 930, 854, 704, np.nan,  np.nan,  109, 38, 0])
y1_75 = np.array([0, 0, 10,       135,     410,     610,     602,   615,  703,  np.nan,   837, 710, 698, 505, np.nan,     175,   95, 33, 0])
y1_50 = np.array([0, 0, 12,       103,  np.nan,     576,  np.nan,   410,  475,  np.nan,   635, 516, 595, 300, np.nan,     122,   60, 25, 0])
y1_25 = np.array([0, 0, 0,         43,  np.nan,     501,     260,   275,  262,  np.nan,   470, 300, 470, 105,     82,      75,   35,  5, 0])
y1_05 = np.array([0, 0, 0,         13,  np.nan,  np.nan,     210,    95,   93,  np.nan,    99, 180,  30,  20,      0,       0,    0,  0, 0])

df_1 = pd.DataFrame({'x' : x1, 'y1_05' : pd.Series(y1_05), 'y1_25' : pd.Series(y1_25), 'y1_50' : pd.Series(y1_50), 
                    'y1_75' : pd.Series(y1_75), 'y1_95' : pd.Series(y1_95)})
df_1 = df_1.set_index(x1)
df_1['y1_05'].interpolate(method='cubic', inplace=True)
df_1['y1_25'].interpolate(method='cubic', inplace=True)
df_1['y1_50'].interpolate(method='cubic', inplace=True)
df_1['y1_75'].interpolate(method='cubic', inplace=True)
df_1['y1_95'].interpolate(method='cubic', inplace=True)

# April 20th data
y2_95 = np.array([0, 39,  197,  380,  532,  703,  823,  987,  1072,  1105,  1103,  1053, 943,  817,  643,  536,  381,  153,  26])
y2_75 = np.array([0, 25,  185,  360,  510,  652,  785,  903,  1001,  1054,  1053,  982,  903,  793,  633,  514,  361,  142,  26])
y2_50 = np.array([0, 15, 172,  320, 490,  623,  740,  850,  930,   975,  973,  942,  863,  760,  610,  480,  350,  130,  15])
y2_25 = np.array([0, 3,  150, 295,   450,   550,   685,  799,  870,  900,  900,  870,  801,   705,  598,  480,  340,  110,  5])
y2_05 = np.array([0, 0, 80, 140,     220,  301,  345,  401,  430,   450, 450,  425,   407,   370,   302,  260,   190,  90,   0])

df_2 = pd.DataFrame({'x' : x2, 'y2_05' : pd.Series(y2_05), 'y2_25' : pd.Series(y2_25), 'y2_50' : pd.Series(y2_50), 
                    'y2_75' : pd.Series(y2_75), 'y2_95' : pd.Series(y2_95)})
df_2 = df_2.set_index(x2)

#y1_05

# setup plot
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(18,7))

# date formatting of x ticks 
#dtFmt = mdates.DateFormatter('%H:%M') # define the formatting
#plt.gca().xaxis.set_major_formatter(dtFmt) 

# add invisible line to set axis
ax[0].fill_between(df_1.x, df_1['y1_95'], df_1['y1_05'], color='red', alpha=0.000001)
ax[1].fill_between(df_2.x, df_2['y2_95'], df_2['y2_05'], color='red', alpha=0.000001)

for hLine in range(200, 1200, 200):
    if(fineTune):
        ax[0].axhline(y= hLine-100, linewidth=0.9, linestyle ="-", color='darkblue', zorder=10)   
    ax[0].axhline(y= hLine, linewidth=0.9, linestyle ="-", color='lightgrey', zorder=10)
    ax[1].axhline(y= hLine, linewidth=0.9, linestyle ="-", color='lightgrey', zorder=10)

# plot 1: add quantiles
ax[0].fill_between(df_1.x, df_1['y1_95'], df_1['y1_05'], color='red', alpha=0.16666, zorder=14)
ax[0].fill_between(df_1.x, df_1['y1_75'], df_1['y1_25'], color='red', alpha=0.3333, zorder=14)
ax[0].fill_between(df_1.x, df_1['y1_50'], df_1['y1_50'], color='#a50000', alpha=1, zorder=14, linewidth=2)

# plot 2: add quantiles
ax[1].fill_between(df_2.x, df_2['y2_95'], df_2['y2_05'], color='red', alpha=0.16666, zorder=14)
ax[1].fill_between(df_2.x, df_2['y2_75'], df_2['y2_25'], color='red', alpha=0.3333, zorder=14)
ax[1].fill_between(df_2.x, df_2['y2_50'], df_2['y2_50'], color='#a50000', alpha=1, zorder=14, linewidth=2)

# plot 1 ticks
ax[0].xaxis.set_major_formatter(mdates.DateFormatter("%H:%M"))
ax[0].tick_params(axis='x', labelsize=16, rotation=45)
ax[0].tick_params(axis='y', labelsize=16)
# plot 2 ticks
ax[1].xaxis.set_major_formatter(mdates.DateFormatter("%H:%M"))
ax[1].tick_params(axis='x', labelsize=16, rotation=45)
ax[1].tick_params(axis='y', labelsize=16)

# plot 1: add text box with date
ax[0].text(0.05, 0.96, '7 April 2020 (Cloudy)', transform=ax[0].transAxes, fontsize=18, verticalalignment='top')
ax[1].text(1.25, 0.96, '20 April 2020 (Sunny)', transform=ax[0].transAxes, fontsize=18, verticalalignment='top')

# y-axis limits set jointly for both sublots
plt.setp(ax, ylim=[0, 1200])

# set rectangular border around both subplots
rect = plt.Rectangle((0.06, -0.15), 0.9, 1.2, fill=True, color="#f9f9f9", lw=1, zorder=-100, transform=fig.transFigure, figure=fig)
fig.patches.extend([rect])
#ax[0].xlabel(fontsize=14)
ax[0].set_ylabel('Global Horizontal Irradiance meters ($W/m^{2}$)', fontsize=20)

# add legend
#plt.legend({'2' : 2}, bbox_to_anchor=(1, 0), loc="lower", bbox_transform=fig.transFigure, ncol=3)
plt.legend({'2' : 2, '3' : 3}, 
           bbox_to_anchor=(-1.0, -0.2, 1.8, 0.2), 
           loc="lower left", 
           mode="expand", 
           borderaxespad=0, 
           ncol=3)
#l5 = plt.legend(bbox_to_anchor=(1, 0), loc="lower right", bbox_transform=fig.transFigure, ncol=3)
#plt.legend([1, 2, 3], ['u','v','u+v'], bbox_to_anchor=(1.04, 1), loc="upper left")

plt.show()
```
