```python
x1 = pd.date_range(start = datetime(2020, 4, 7, 6,   0), 
                   end   = datetime(2020, 4, 7, 19, 30), 
                   freq  = '45min')

x2 = pd.date_range(start = datetime(2020, 4, 20, 6,   0), 
                   end   = datetime(2020, 4, 20, 19, 30), 
                   freq  = '45min')

y1_95 = np.array([0, 0, 20,    np.nan,  np.nan,     655,  np.nan,   805,  902,  np.nan,  1003, 930, 854, 704, np.nan,  np.nan,  109, 38, 0])
y1_75 = np.array([0, 0, 10,       135,     410,     610,     602,   615,  703,  np.nan,   837, 710, 698, 505, np.nan,     175,   95, 33, 0])
y1_50 = np.array([0, 0, 12,       103,  np.nan,     596,  np.nan,   410,  475,  np.nan,   635, 516, 595, 300, np.nan,     122,   60, 25, 0])
y1_25 = np.array([0, 0, 0,         43,  np.nan,     501,     260,   275,  262,  np.nan,   470, 300, 470, 105,     82,      75,   35,  5, 0])
y1_05 = np.array([0, 0, 0,         13,  np.nan,  np.nan,     210,    95,   93,  np.nan,    99, 180,  30,  20,      0,       0,    0,  0, 0])

df_1 = pd.DataFrame({'x' : x1, 'y1_05' : pd.Series(y1_05), 'y1_25' : pd.Series(y1_25), 'y1_50' : pd.Series(y1_50), 
                    'y1_75' : pd.Series(y1_75), 'y1_95' : pd.Series(y1_95)})
df_1 = df_1.set_index(x1)
df_1['y1_05'].interpolate(method='cubic', inplace=True)
df_1['y1_25'].interpolate(method='cubic', inplace=True)
df_1['y1_50'].interpolate(method='cubic', inplace=True)
df_1['y1_75'].interpolate(method='cubic', inplace=True)
df_1['y1_95'].interpolate(method='cubic', inplace=True)

#y1_05

# setup plot
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(18,7))

# date formatting of x ticks 
dtFmt = mdates.DateFormatter('%H:%M') # define the formatting
plt.gca().xaxis.set_major_formatter(dtFmt) 

# add invisible line to set axis
ax[0].fill_between(df_1.x, df_1['y1_95'], df_1['y1_05'], color='red', alpha=0.000001)

# add horizontal lines
fineTune = False

for hLine in range(200, 1200, 200):
    if(fineTune):
        ax[0].axhline(y= hLine-100, linewidth=0.9, linestyle ="-", color='darkblue', zorder=10)   
    ax[0].axhline(y= hLine, linewidth=0.9, linestyle ="-", color='lightgrey', zorder=10)

# add quantiles
ax[0].fill_between(df_1.x, df_1['y1_95'], df_1['y1_05'], color='red', alpha=0.16666, zorder=14)
ax[0].fill_between(df_1.x, df_1['y1_75'], df_1['y1_25'], color='red', alpha=0.3333, zorder=14)

# show every 12th tick on x axes
ax[0].set_major_locator(mdates.MonthLocator(interval=1))
ax[0].xticks(rotation=90, fontweight='light',  fontsize='x-small',)

# y-axis limits set jointly for both sublots
plt.setp(ax, ylim=[0, 1200])

# set rectangular border around both subplots
rect = plt.Rectangle((0.06, 0.02), 0.9, 1.0, fill=True, color="#f9f9f9", lw=1, zorder=-100, transform=fig.transFigure, figure=fig)
fig.patches.extend([rect])
#ax[0].xlabel(fontsize=14)
ax[0].set_ylabel('Global Horizontal Irradiance meters ($W/m^{2}$)', fontsize=18)

plt.show()
```
